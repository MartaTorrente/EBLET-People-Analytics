## 1. 🏢 Generación de LegacyTech Solutions

Generamos internamente una empresa ficticia con perfil de **Riesgo Boreout**:
- Empresa tech madura (15 años)
- Cultura real: **Jerárquica**
- 180 empleados
- Boreout alto, Burnout bajo
- 72 preguntas (incluye CVF)

```python
# =====================================================
# IMPORTS Y CONFIGURACIÓN
# =====================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

import os
import sys

RAIZ_PROYECTO = r"C:\Users\torre\OneDrive\Escritorio\EBLET-People-Analytics\Python"
sys.path.insert(0, os.path.join(RAIZ_PROYECTO, "src"))

from scores import calcular_kpis_empleado, calcular_kpis_empresa, clasificar_escenario_empresa
from costes_rotacion import (
    calcular_coste_rotacion_empleado,
    calcular_costes_empresa,
    calcular_roi_intervencion
)

# Paleta de colores consistente con otros notebooks
COLORES_ESCENARIOS = {
    "saludable": "#2ecc71",
    "estable": "#f1c40f",
    "riesgo_burnout": "#e67e22",
    "riesgo_boreout": "#3498db",
    "critico": "#e74c3c"
}

ETIQUETAS_ES = {
    "saludable": "🟢 Saludable",
    "estable": "🟡 Estable",
    "riesgo_burnout": "🟠 Riesgo Burnout",
    "riesgo_boreout": "🔵 Riesgo Boreout",
    "critico": "🔴 Crítico"
}

print("✅ Notebook 5 cargado correctamente")
```

## 1. 📥 Carga de la Empresa Demo

```python
# =====================================================
# CARGA DE EMPRESA DEMO
# =====================================================

ruta_demo = os.path.join(RAIZ_PROYECTO, "datasets/empresa_demo.csv")
df_demo = pd.read_csv(ruta_demo)

print("="*70)
print("🏢 EMPRESA DEMO: LEGACYTECH SOLUTIONS")
print("="*70)
print(f"\n👥 Empleados: {len(df_demo)}")
print(f"🏭 Departamentos: {df_demo['departamento'].nunique()}")
print(f"📊 Seniority levels: {df_demo['seniority'].nunique()}")

print("\n📋 Distribución por departamento:")
print(df_demo['departamento'].value_counts().to_string())

print("\n👤 Distribución por seniority:")
print(df_demo['seniority'].value_counts().to_string())

print("\n💼 Modalidad de trabajo:")
print(df_demo['modalidad'].value_counts().to_string())
```

## 2. 📊 Cálculo de KPIs por Empleado

```python
# =====================================================
# CÁLCULO DE KPIs
# =====================================================

df_demo_kpis = calcular_kpis_empleado(df_demo)

# Calcular coste individual
df_demo_kpis["coste_rotacion_individual"] = df_demo_kpis.apply(
    calcular_coste_rotacion_empleado, axis=1
)

# KPIs medios de la empresa
kpi_medios = {
    "Burnout": df_demo_kpis["kpi_burnout"].mean(),
    "Boreout": df_demo_kpis["kpi_boreout"].mean(),
    "Bienestar": df_demo_kpis["kpi_bienestar"].mean(),
    "Rotación": df_demo_kpis["kpi_rotacion"].mean(),
    "Contexto": df_demo_kpis["kpi_contexto"].mean()
}

print("="*70)
print("📈 KPIs MEDIOS DE LEGACYTECH SOLUTIONS")
print("="*70)
for kpi, valor in kpi_medios.items():
    print(f"   {kpi:15s}: {valor:.2f}")

print(f"\n💰 Coste medio de rotación por empleado: {df_demo_kpis['coste_rotacion_individual'].mean():,.2f}€")
print(f"💸 Coste total estimado: {df_demo_kpis['coste_rotacion_individual'].sum():,.2f}€")
```

## 3. 🎯 Clasificación de la Empresa

```python
# =====================================================
# CLASIFICACIÓN DE LA EMPRESA
# =====================================================

# Crear DataFrame a nivel empresa
df_empresa_demo = pd.DataFrame({
    "empresa_id": ["DEMO_001"],
    "kpi_burnout": [kpi_medios["Burnout"]],
    "kpi_boreout": [kpi_medios["Boreout"]],
    "kpi_bienestar": [kpi_medios["Bienestar"]],
    "kpi_rotacion": [kpi_medios["Rotación"]],
    "kpi_contexto": [kpi_medios["Contexto"]]
})

# Clasificar
df_empresa_clasificada = clasificar_escenario_empresa(df_empresa_demo)
escenario_detectado = df_empresa_clasificada["escenario_predicho"].iloc[0]

print("="*70)
print("🎯 CLASIFICACIÓN DE LA EMPRESA")
print("="*70)
print(f"\n✅ Escenario detectado: {ETIQUETAS_ES[escenario_detectado]}")
print(f"\n📊 KPIs vs Umbrales:")
print(f"   Burnout:   {kpi_medios['Burnout']:.2f} (umbral alto: 3.86)")
print(f"   Boreout:   {kpi_medios['Boreout']:.2f} (umbral alto: 3.00) ⚠️ SUPERADO")
print(f"   Bienestar: {kpi_medios['Bienestar']:.2f} (umbral bajo: 2.60) ⚠️ POR DEBAJO")
```

## 4. 📡 Comparación con el Benchmark

```python
# =====================================================
# CARGAR BENCHMARK (5 ESCENARIOS)
# =====================================================

ESCENARIOS = ["saludable", "estable", "riesgo_burnout", "riesgo_boreout", "critico"]

dfs_benchmark = {}
for esc in ESCENARIOS:
    ruta = os.path.join(RAIZ_PROYECTO, f"datasets/{esc}/empleados.csv")
    df = pd.read_csv(ruta)
    df["escenario"] = esc
    dfs_benchmark[esc] = df

# KPIs medios del benchmark
benchmark_kpis = {}
for esc in ESCENARIOS:
    benchmark_kpis[esc] = {
        "Burnout": dfs_benchmark[esc]["kpi_burnout"].mean(),
        "Boreout": dfs_benchmark[esc]["kpi_boreout"].mean(),
        "Bienestar": dfs_benchmark[esc]["kpi_bienestar"].mean(),
        "Rotación": dfs_benchmark[esc]["kpi_rotacion"].mean(),
        "Contexto": dfs_benchmark[esc]["kpi_contexto"].mean()
    }

df_benchmark = pd.DataFrame(benchmark_kpis).T
df_benchmark.index.name = "escenario"

print("="*70)
print("📡 BENCHMARK DE LOS 5 ESCENARIOS")
print("="*70)
print(df_benchmark.round(2))
```

## 5. 🕸️ Visualización Radar: Empresa vs Benchmark

```python
# =====================================================
# RADAR CHART: EMPRESA vs BENCHMARK
# =====================================================

# Preparar datos para el radar
kpi_nombres = ["Bienestar", "Contexto", "Sin Burnout", "Sin Boreout", "Retención"]

# Valores de la empresa (invertir los negativos para que mayor = mejor)
valores_empresa = [
    kpi_medios["Bienestar"],
    kpi_medios["Contexto"],
    5 - kpi_medios["Burnout"],
    5 - kpi_medios["Boreout"],
    5 - kpi_medios["Rotación"]
]

# Valores del benchmark del escenario detectado
valores_benchmark = [
    benchmark_kpis[escenario_detectado]["Bienestar"],
    benchmark_kpis[escenario_detectado]["Contexto"],
    5 - benchmark_kpis[escenario_detectado]["Burnout"],
    5 - benchmark_kpis[escenario_detectado]["Boreout"],
    5 - benchmark_kpis[escenario_detectado]["Rotación"]
]

# Valores del escenario "Saludable" como referencia ideal
valores_saludable = [
    benchmark_kpis["saludable"]["Bienestar"],
    benchmark_kpis["saludable"]["Contexto"],
    5 - benchmark_kpis["saludable"]["Burnout"],
    5 - benchmark_kpis["saludable"]["Boreout"],
    5 - benchmark_kpis["saludable"]["Rotación"]
]

# Cerrar el polígono
valores_empresa_cerrados = valores_empresa + [valores_empresa[0]]
valores_benchmark_cerrados = valores_benchmark + [valores_benchmark[0]]
valores_saludable_cerrados = valores_saludable + [valores_saludable[0]]
labels_cerrados = kpi_nombres + [kpi_nombres[0]]

# Colores explícitos
color_empresa = COLORES_ESCENARIOS[escenario_detectado]

fig = go.Figure()

# 1. Empresa demo (protagonista) - AZUL
fig.add_trace(go.Scatterpolar(
    r=valores_empresa_cerrados,
    theta=labels_cerrados,
    fill='toself',
    name='🏢 LegacyTech Solutions',  # ← Nombre claro
    line_color=color_empresa,
    fillcolor=color_empresa,
    opacity=0.7,
    line=dict(width=3)
))

# 2. Benchmark del escenario - GRIS
fig.add_trace(go.Scatterpolar(
    r=valores_benchmark_cerrados,
    theta=labels_cerrados,
    fill='toself',
    name=f'📊 Benchmark {ETIQUETAS_ES[escenario_detectado]}',  # ← Nombre claro
    line_color='gray',
    fillcolor='rgba(128, 128, 128, 0.2)',
    line_dash='dash',
    opacity=0.5
))

# 3. Escenario saludable como referencia - VERDE
fig.add_trace(go.Scatterpolar(
    r=valores_saludable_cerrados,
    theta=labels_cerrados,
    fill='toself',
    name='🟢 Ideal (Saludable)',  # ← Nombre claro
    line_color='#2ecc71',
    fillcolor='rgba(46, 204, 113, 0.1)',
    line_dash='dot',
    opacity=0.4
))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 5])),
    title=f"🕸️ LegacyTech vs Benchmark - Escenario: {ETIQUETAS_ES[escenario_detectado]}",
    showlegend=True,
    height=600,
    width=800,
    legend=dict(
        orientation="v",
        yanchor="top",
        y=1.1,
        xanchor="left",
        x=0
    )
)
fig.show()

print(f"""
💡 INTERPRETACIÓN:
- Cuanto más grande el área, más saludable la organización
- La empresa (azul) se clasifica como: {ETIQUETAS_ES[escenario_detectado]}
- Se asemeja al benchmark de {ETIQUETAS_ES[escenario_detectado]} (gris)
- Comparado con el ideal (verde), hay una brecha significativa en:
  * "Sin Boreout" → Principal déficit de la empresa
  * "Retención" → Alta intención de rotación
  * "Bienestar" → Bajo bienestar general
- El área "Sin Burnout" es relativamente buena (burnout bajo = 2.58)
  Esto confirma el diagnóstico: el problema es BOREOUT, no burnout
""")
```

## 7. 💰 Impacto Económico

```python
# =====================================================
# ANÁLISIS DE COSTES
# =====================================================

coste_total = df_demo_kpis["coste_rotacion_individual"].sum()
coste_medio = df_demo_kpis["coste_rotacion_individual"].mean()
masa_salarial = df_demo_kpis["salario"].sum()
pct_masa_salarial = (coste_total / masa_salarial) * 100

# Tasa de rotación estimada
def kpi_a_tasa(kpi):
    if kpi <= 1.5: return 0.05
    elif kpi <= 2.5: return 0.10
    elif kpi <= 3.5: return 0.20
    elif kpi <= 4.5: return 0.35
    else: return 0.50

tasa_rotacion = kpi_a_tasa(kpi_medios["Rotación"])
n_bajas_estimadas = int(len(df_demo_kpis) * tasa_rotacion)

print("="*70)
print("💰 IMPACTO ECONÓMICO - LEGACYTECH SOLUTIONS")
print("="*70)
print(f"\n👥 Empleados: {len(df_demo_kpis)}")
print(f"💵 Masa salarial anual: {masa_salarial:,.2f}€")
print(f"\n💸 COSTES DE ROTACIÓN:")
print(f"   Coste total estimado: {coste_total:>12,.2f}€")
print(f"   Coste medio por empleado: {coste_medio:>8,.2f}€")
print(f"   % sobre masa salarial: {pct_masa_salarial:.1f}%")
print(f"\n📉 ROTACIÓN:")
print(f"   KPI Rotación: {kpi_medios['Rotación']:.2f}")
print(f"   Tasa anual estimada: {tasa_rotacion*100:.0f}%")
print(f"   Bajas estimadas al año: {n_bajas_estimadas} empleados")

# Tarjetas visuales
fig = make_subplots(
    rows=1, cols=4,
    subplot_titles=(
        f"💸 Coste Total<br><b>{coste_total/1e6:.2f}M€</b>",
        f"💵 Coste/Empleado<br><b>{coste_medio/1e3:.1f}K€</b>",
        f"📊 % Masa Salarial<br><b>{pct_masa_salarial:.1f}%</b>",
        f"📉 Bajas/Año<br><b>{n_bajas_estimadas}</b>"
    ),
    specs=[[{"type": "indicator"}]*4]
)

fig.add_trace(go.Indicator(
    mode="number",
    value=coste_total/1e6,
    number={'suffix': "M€", 'font': {'size': 30}},
), row=1, col=1)

fig.add_trace(go.Indicator(
    mode="number",
    value=coste_medio/1e3,
    number={'suffix': "K€", 'font': {'size': 30}},
), row=1, col=2)

fig.add_trace(go.Indicator(
    mode="number",
    value=pct_masa_salarial,
    number={'suffix': "%", 'font': {'size': 30}},
), row=1, col=3)

fig.add_trace(go.Indicator(
    mode="number",
    value=n_bajas_estimadas,
    number={'font': {'size': 30}},
), row=1, col=4)

fig.update_layout(
    title="💰 Tarjetas de Impacto Económico",
    height=200
)
fig.show()
```
