# 👤 EBLET-Lite: Procesamiento y Análisis de Respuestas Reales

## 🎯 Objetivo

Procesar las respuestas obtenidas mediante EBLET-Lite para calcular los principales indicadores de bienestar laboral, clasificar el perfil individual de cada participante y generar un informe comparativo respecto al benchmark desarrollado en este proyecto.

## 📋 EBLET-Lite: Versión Abreviada de Cribado

EBLET-Lite es una **versión reducida** de EBLET diseñada para aplicación rápida individual.


## 🧩 Flujo de procesamiento

El análisis sigue las siguientes etapas:

1. Importación de las respuestas obtenidas mediante Google Forms.
2. Preparación y validación de los datos.
3. Cálculo de los KPIs individuales.
4. Clasificación automática del perfil de bienestar.
5. Comparación con el benchmark sintético desarrollado en este proyecto.
6. Generación de un informe individual con los resultados.

### ⚠️ Limitaciones de EBLET-Lite

EBLET-Lite es una **versión abreviada de cribado individual** basada en las dimensiones centrales de EBLET Enterprise. Mantiene los constructos principales de bienestar laboral, pero reduce la profundidad diagnóstica para permitir una aplicación rápida (~5 minutos).

**Dimensiones no incluidas:**
- Recursos y demandas laborales (JD-R)
- Autoeficacia
- Satisfacción laboral completa
- Eficacia profesional
- Infraocupación completa

Por tanto, EBLET-Lite **evalúa indicadores relacionados con agotamiento y desconexión laboral**, En consecuencia, EBLET-Lite debe entenderse como una herramienta de cribado orientada a la identificación temprana de posibles situaciones de riesgo, no como un instrumento de evaluación psicológica exhaustiva.

## 1. KPIs según variables demográficas y laborales

Este gráfico compara los niveles medios de **burnout**, **boreout**, **bienestar** e **intención de cambio** según distintas características demográficas y laborales.

El análisis permite detectar diferencias entre grupos y localizar perfiles que pueden presentar un mayor riesgo de desgaste, desmotivación o rotación.

Los resultados deben interpretarse de forma descriptiva, especialmente en aquellas categorías con pocos participantes.

![KPIs según variables demográficas y laborales](../visualizaciones/03_kpis_segun_variables_demograficas.png)



- Se observan diferencias en los niveles medios de los KPIs entre algunos grupos demográficos y laborales.
- Los mayores niveles de burnout y boreout no siempre se concentran en los mismos colectivos, lo que sugiere que ambos fenómenos responden a dinámicas laborales diferentes.
- El bienestar presenta, en general, un comportamiento inverso al burnout y al boreout.
- La intención de cambio tiende a ser mayor en aquellos grupos con peores indicadores de bienestar laboral.

> **Nota:** Las diferencias observadas son descriptivas y deben interpretarse con cautela, especialmente en categorías con un número reducido de participantes.

## 2. Distribución de perfiles de bienestar

Tras analizar los KPIs por grupos demográficos y laborales, el siguiente paso consiste en conocer cómo se distribuyen los participantes según los perfiles definidos por el modelo EBLET.

Cada perfil representa una combinación de los niveles de **burnout**, **boreout** y **bienestar**, permitiendo clasificar a cada persona en función de su situación laboral predominante.

![Distribución de perfiles de bienestar](../visualizaciones/04_distribucion_perfiles_bienestar.png)



- La distribución de perfiles ofrece una visión global del estado de bienestar de la muestra.
- Los perfiles **Equilibrio** y **Estable** representan a las personas con una situación laboral más favorable.
- Los perfiles de **Riesgo de Burnout**, **Riesgo de Boreout** y **Crítico** identifican colectivos que podrían requerir actuaciones específicas de prevención o intervención.
- El peso relativo de cada perfil permite estimar la proporción de trabajadores que se encuentran en situaciones de mayor vulnerabilidad dentro de la organización.

> **Nota:** La clasificación de perfiles se basa en los indicadores calculados por el modelo EBLET y constituye una herramienta descriptiva para segmentar el estado de bienestar de la muestra.

## 3. Distribución de culturas organizacionales percibidas

Una vez analizada la distribución de los perfiles de bienestar, resulta de interés conocer cuál es la cultura organizacional predominante percibida por los participantes.

La clasificación se basa en el modelo **Competing Values Framework (CVF)**, que distingue cuatro tipologías culturales: **Clan**, **Adhocracia**, **Mercado** y **Jerárquica**.

![Distribución de culturas organizacionales](../visualizaciones/05_distribucion_culturas_organizacionales.png)


- La distribución muestra qué tipo de cultura es percibida con mayor frecuencia dentro de la organización.
- Cada cultura refleja diferentes estilos de liderazgo, formas de trabajo y prioridades organizacionales.
- La predominancia de una determinada cultura puede ayudar a interpretar posteriormente las diferencias observadas en bienestar, burnout, boreout e intención de cambio.
- Este análisis proporciona el contexto organizacional sobre el que se desarrollan las experiencias laborales de los participantes.

> **Nota:** La cultura dominante se asigna a partir de la dimensión cultural con mayor puntuación obtenida por cada participante según el modelo CVF.

## 4. Relación entre cultura organizacional y KPIs

Una vez identificadas las culturas organizacionales percibidas, se comparan los niveles medios de **burnout**, **boreout**, **bienestar** e **intención de cambio** en cada una de ellas.

Este análisis permite observar si determinados contextos culturales se asocian con mejores o peores indicadores de bienestar laboral.

![Relación entre cultura organizacional y KPIs](../visualizaciones/06_relacion_cultura_kpis.png)



- Los niveles medios de los KPIs varían entre las distintas culturas organizacionales percibidas.
- Algunas culturas presentan mayores niveles de bienestar, mientras que otras concentran valores más elevados de burnout, boreout o intención de cambio.
- La comparación permite identificar qué entornos culturales parecen estar asociados con situaciones laborales más favorables o más vulnerables.
- Estos resultados ayudan a contextualizar los perfiles de bienestar observados anteriormente.

> **Nota:** Las diferencias mostradas son descriptivas y no implican una relación causal directa entre la cultura organizacional y los indicadores analizados.

## 5. Mapa de posicionamiento individual: Burnout vs. Boreout

El mapa representa la posición de cada participante según sus niveles de **burnout** y **boreout**, permitiendo visualizar simultáneamente ambos fenómenos y clasificar a las personas en cuatro zonas de riesgo.

Los umbrales establecidos dividen el espacio en las categorías **Equilibrio/Estable**, **Riesgo de Burnout**, **Riesgo de Boreout** y **Crítico**, facilitando la identificación de perfiles individuales y la detección de posibles situaciones de vulnerabilidad.




- El gráfico permite identificar visualmente la distribución individual de los participantes según sus niveles de burnout y boreout.
- La concentración de puntos en cada cuadrante ofrece una visión rápida de la proporción de personas en situaciones de equilibrio o de riesgo.
- Los participantes situados en la zona **Crítica** presentan simultáneamente niveles elevados de burnout y boreout, constituyendo el colectivo con mayor vulnerabilidad.
- Esta representación facilita la identificación de casos individuales y complementa el análisis realizado mediante los perfiles EBLET.

> **Nota:** Los desplazamientos aplicados a algunos puntos tienen únicamente un propósito visual para evitar la superposición de participantes con puntuaciones idénticas; las puntuaciones originales no se modifican.
