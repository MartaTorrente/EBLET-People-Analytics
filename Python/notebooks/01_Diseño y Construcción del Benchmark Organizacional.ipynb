## Objetivo

En este notebook se describe la metodología empleada para construir el **Benchmark Organizacional EBLET**, una referencia sintética diseñada para contextualizar los indicadores obtenidos mediante la encuesta de bienestar laboral.

A lo largo del notebook se presentan las diferentes etapas del proceso de generación del benchmark, desde el diseño conceptual del modelo hasta el cálculo de los indicadores organizacionales y la estimación del coste de rotación.

El resultado es una base de referencia que permitirá comparar organizaciones con distintos niveles de salud organizacional, sirviendo como punto de partida para el análisis del caso práctico desarrollado en el Notebook 03.

# 1. ¿Por qué construir un benchmark?

El proyecto EBLET no parte de un conjunto de datos organizacionales reales que pueda utilizarse como referencia. Por ello, fue necesario construir un benchmark propio que permitiera contextualizar los resultados obtenidos mediante la encuesta.

Este benchmark se diseñó para representar diferentes escenarios de salud organizacional mediante la simulación de empresas y empleados con características diversas. Su finalidad no es establecer un estándar universal, sino disponer de una referencia coherente dentro del propio proyecto que permita comparar organizaciones bajo un mismo marco metodológico.

En el siguiente notebook, este benchmark se utilizará como base para evaluar el caso práctico de **LegacyTech**.

# 2. Diseño conceptual del benchmark

El benchmark organizacional de EBLET se construye mediante un proceso de simulación que reproduce las principales etapas presentes en una organización, desde la definición de su contexto hasta la obtención de indicadores que permiten evaluar su estado de salud organizacional.

A diferencia de una generación aleatoria de respuestas, el modelo establece una relación entre las características de la organización, las características individuales de los empleados y los estados psicológicos que experimentan. Estos estados se traducen posteriormente en las respuestas de la encuesta y, a partir de ellas, se calculan los indicadores utilizados para construir el benchmark.

Este enfoque permite generar organizaciones coherentes entre sí y representativas de diferentes escenarios, proporcionando una base sólida para la comparación y el análisis posterior.

> **Idea clave**
>
> El benchmark no se genera directamente a partir de respuestas aleatorias, sino mediante un proceso secuencial que relaciona el contexto organizacional con el comportamiento individual y los indicadores finales.

# 4. Definición de los escenarios organizacionales

El punto de partida del benchmark consiste en la definición de distintos escenarios organizacionales que representan diferentes niveles de salud organizacional.

Cada escenario establece un conjunto de parámetros que condicionan el entorno laboral de las empresas simuladas y, de forma indirecta, el comportamiento de los empleados durante el proceso de generación de datos.

Estos escenarios no representan empresas concretas, sino perfiles organizacionales diseñados para cubrir un amplio rango de situaciones que pueden encontrarse en el ámbito laboral.

| Escenario         | Descripción                                                                   |
| ----------------- | ----------------------------------------------------------------------------- |
| 🟢 Saludable      | Alto bienestar, bajo burnout y boreout, entorno favorable.                    |
| 🔵 Estable        | Situación equilibrada con margen de mejora.                                   |
| 🟠 Riesgo Burnout | Elevada carga de trabajo y agotamiento.                                       |
| 🟣 Riesgo Boreout | Baja estimulación, infraocupación y desmotivación.                            |
| 🔴 Crítico        | Deterioro generalizado del entorno laboral y de los indicadores psicológicos. |


Nota: Estos escenarios no representan empresas reales, sino perfiles organizacionales diseñados para generar un benchmark consistente y cubrir diferentes situaciones que pueden encontrarse en el ámbito laboral.

| Escenario      | Burnout | Boreout | Bienestar |     Contexto     |
| -------------- | :-----: | :-----: | :-------: | :--------------: |
| Saludable      |   Bajo  |   Bajo  |    Alto   |     Favorable    |
| Estable        |  Medio  |   Bajo  |   Medio   |      Neutro      |
| Riesgo Burnout |   Alto  |   Bajo  |    Bajo   |     Exigente     |
| Riesgo Boreout |   Bajo  |   Alto  |    Bajo   | Poco estimulante |

🟢 Saludable → 🔵 Estable → 🟠 Riesgo → 🔴 Crítico

Cada empresa del benchmark se genera a partir de uno de estos escenarios, que actúa como punto de partida para definir tanto las características organizacionales como la distribución de las variables psicológicas de sus empleados.

# 5. Generación de empresas

Una vez definidos los escenarios organizacionales, el siguiente paso consiste en generar las empresas que formarán parte del benchmark.

Cada empresa representa una organización simulada cuyas características se establecen a partir del escenario al que pertenece. De este modo, organizaciones clasificadas dentro de un mismo escenario comparten un comportamiento general, aunque mantienen suficiente variabilidad para evitar que dos empresas sean idénticas.

Para cada empresa se generan diferentes atributos organizacionales que servirán posteriormente para modelar el comportamiento de sus empleados y calcular los indicadores finales.

> **Idea clave**
>
> El escenario organizacional actúa como punto de partida para definir el contexto de cada empresa, pero cada organización incorpora variabilidad propia que la hace única dentro del benchmark.

| Variable                         | Descripción                                                   |
| -------------------------------- | ------------------------------------------------------------- |
| **ID de empresa**                | Identificador único de la organización.                       |
| **Escenario**                    | Saludable, Estable, Riesgo Burnout, Riesgo Boreout o Crítico. |
| **Sector**                       | Actividad económica de la empresa.                            |
| **Ciudad**                       | Localización geográfica simulada.                             |
| **Tamaño**                       | Número de empleados asignados.                                |
| **Cultura organizacional (CVF)** | Perfil cultural predominante de la empresa.                   |


Aunque las empresas pertenecen a un escenario determinado, ninguno de sus atributos se genera de forma completamente determinista. La incorporación de variabilidad controlada permite obtener un benchmark más realista y representar organizaciones con características diversas dentro de un mismo escenario.

# 6. Generación de empleados

Una vez definidas las características de cada empresa, se generan los empleados que formarán parte de cada organización simulada.

Cada empleado hereda el contexto organizacional de la empresa a la que pertenece, pero dispone de características individuales propias que introducen variabilidad en el modelo. De este modo, dos empleados que trabajan en la misma empresa pueden experimentar de forma diferente una misma situación laboral.

Estas características individuales constituyen la base sobre la que posteriormente se modelarán los estados psicológicos y las respuestas a la encuesta.

> **Idea clave**
>
> Los empleados comparten un mismo entorno organizacional, pero cada uno presenta características personales que hacen que su experiencia laboral sea única.

| Variable | Descripción |
|----------|-------------|
| **ID empleado** | Identificador único del empleado. |
| **Edad** | Edad simulada dentro de un rango laboral realista. |
| **Género** | Distribución simulada de la plantilla. |
| **Departamento** | Área funcional a la que pertenece. |
| **Antigüedad** | Tiempo de permanencia en la empresa. |
| **Nivel profesional** | Categoría o seniority del empleado. |
| **Modalidad de trabajo** | Presencial, híbrido o remoto. |
| **Salario** | Retribución estimada según el nivel profesional. |

La combinación de estas variables permite generar plantillas heterogéneas y más próximas a la realidad organizacional. Esta diversidad resulta esencial para evitar organizaciones excesivamente homogéneas y reproducir la variabilidad que existe entre los empleados de una misma empresa.

Hasta este punto, cada empleado dispone únicamente de características demográficas y laborales. Sin embargo, estas variables por sí solas no determinan las respuestas de la encuesta.

En la siguiente etapa del proceso, el modelo incorpora una dimensión psicológica que permite simular cómo el contexto organizacional y las características individuales influyen en el bienestar, el burnout, el boreout y la intención de rotación de cada empleado.

Esta composición permite introducir diferencias realistas en variables relacionadas, como el salario, la experiencia profesional, la antigüedad y el coste potencial de rotación.

# 7. Modelo psicológico

Hasta este punto del proceso, cada empleado dispone de unas características demográficas y laborales. Sin embargo, estas variables no determinan por sí solas cómo percibe su trabajo o cómo responderá a la encuesta.

Para representar esta dimensión subjetiva, el modelo incorpora un conjunto de **estados psicológicos latentes**, es decir, variables que no pueden observarse directamente, pero que influyen en el comportamiento y en las respuestas de cada empleado.

> **¿Qué es una variable latente?**
>
> Una variable latente representa un estado o característica psicológica que no puede medirse de forma directa, sino que se infiere a partir de indicadores observables. En este proyecto, estos estados se utilizan para generar respuestas coherentes a la encuesta y reproducir diferentes perfiles organizacionales.

En EBLET se modelan cuatro estados psicológicos principales:

- **Burnout**, asociado al agotamiento y al estrés laboral.
- **Boreout**, relacionado con la infraocupación y la falta de estímulo.
- **Bienestar**, que refleja el estado general de salud y satisfacción laboral.
- **Intención de rotación**, que representa la predisposición del empleado a abandonar la organización.

Estos estados no se generan de forma aleatoria. Cada uno de ellos depende tanto del contexto organizacional definido por el escenario de la empresa como de las características individuales de cada empleado.

De este modo, dos empleados pertenecientes a una misma organización pueden experimentar niveles diferentes de burnout o bienestar, reproduciendo la variabilidad que existe en cualquier entorno laboral real.

## 10. Modelo de Cultura Organizacional

Además de los indicadores de bienestar y riesgos psicosociales, EBLET incorpora la evaluación de la cultura organizacional como una dimensión clave para interpretar el funcionamiento de las empresas.

Para ello se adopta el **Competing Values Framework (CVF)**, uno de los modelos de cultura organizacional más utilizados en investigación y consultoría. Este modelo clasifica la cultura de una organización en cuatro tipologías principales, cada una asociada a diferentes estilos de liderazgo, formas de trabajo y prioridades estratégicas.

> 💡 **Idea clave**
>
> La cultura organizacional no se interpreta como un indicador positivo o negativo, sino como un conjunto de características que condicionan la forma en que las personas trabajan y se relacionan dentro de la organización.

Las cuatro dimensiones consideradas son:

| Cultura | Características principales |
|----------|-----------------------------|
| 🤝 Clan | Colaboración, confianza, trabajo en equipo y desarrollo de las personas. |
| 🚀 Adhocracia | Innovación, creatividad, flexibilidad y capacidad de adaptación. |
| 🎯 Mercado | Orientación a resultados, competitividad y consecución de objetivos. |
| 🏛️ Jerarquía | Estabilidad, normas, planificación y control de los procesos. |

En el proceso de simulación, cada empresa recibe un perfil cultural dominante coherente con el escenario organizacional definido. Posteriormente, dicho perfil es estimado a partir de las respuestas del cuestionario mediante los ítems específicos del modelo CVF.

De este modo, la cultura organizacional pasa a formar parte del conjunto de indicadores utilizados para caracterizar cada organización dentro del benchmark.

# 12. Estimación del coste de rotación

Además de los indicadores relacionados con el bienestar laboral, el benchmark incorpora una estimación del impacto económico asociado a la rotación de personal.

El objetivo de esta estimación es trasladar los resultados del análisis psicosocial a un contexto de negocio, permitiendo cuantificar las posibles consecuencias económicas derivadas de la pérdida de talento.

En EBLET, el coste de rotación se estima de forma individual para cada empleado considerando diferentes factores relacionados con su perfil profesional y su retribución. Posteriormente, estos costes se agregan para obtener una estimación del impacto económico potencial de cada organización.

> **Idea clave**
>
> El coste de rotación no representa un gasto real observado, sino una estimación basada en distintos factores que permite comparar el impacto potencial de la rotación entre organizaciones.

| Factor considerado                | Influencia en el coste                                                   |
| --------------------------------- | ------------------------------------------------------------------------ |
| Salario                           | Determina el coste base de sustitución.                                  |
| Nivel profesional                 | Empleados de mayor responsabilidad implican un mayor coste de reemplazo. |
| Incorporación y formación         | Coste asociado al proceso de adaptación del nuevo empleado.              |
| Pérdida temporal de productividad | Impacto durante el periodo hasta recuperar el rendimiento esperado.      |
| Probabilidad de rotación          | Ajusta el coste potencial en función del riesgo estimado de salida.      |


La incorporación de esta métrica permite complementar los indicadores psicológicos con una estimación de su posible impacto económico, facilitando la priorización de acciones de mejora desde una perspectiva tanto organizacional como empresarial.

### Impacto económico de la rotación

Además de los indicadores psicológicos, el benchmark incorpora una estimación del impacto económico asociado a la intención de salida de los empleados. El coste esperado se obtiene combinando el salario, el factor de reemplazo correspondiente al nivel de seniority y la probabilidad estimada de salida derivada del KPI de rotación.
