# 📋 Encuesta de Bienestar Laboral y Experiencia en el Trabajo (EBLET)

**Versión:** 1.0  
**Duración estimada:** 10-15 minutos  
**Total de preguntas:** 48 (numeradas) + 12 (demográficas)

---

## 📝 Instrucciones Generales

Esta encuesta tiene como objetivo evaluar el bienestar laboral y la experiencia en el trabajo. Tus respuestas son **completamente anónimas y confidenciales**. No hay respuestas correctas o incorrectas; simplemente indica tu grado de acuerdo o frecuencia según tu experiencia personal.

**Escala de respuesta:** Cuando se indique "Escala 1-5", utiliza la siguiente referencia:

| Valor | Significado |
|-------|-------------|
| **1** | Totalmente en desacuerdo / Nunca |
| **2** | En desacuerdo / Raramente |
| **3** | Neutral / A veces |
| **4** | De acuerdo / Frecuentemente |
| **5** | Totalmente de acuerdo / Siempre |
