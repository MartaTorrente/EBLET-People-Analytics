#  Planificación del Proyecto EBLET - People Analytics Framework


## MISIÓN

Desarrollar un **framework de People Analytics** de referencia que contenga las herramientas necesarias para:

- ✅ Evaluar el bienestar laboral mediante una encuesta estructurada (EBLET)
- ✅ Medir, controlar y prevenir el **burnout** y el **boreout**
- ✅ Calcular indicadores de **bienestar** e **intención de rotación**
- ✅ Generar diagnósticos organizacionales accionables para RRHH
- ✅ Proporcionar benchmarks comparativos para ubicar organizaciones en un espectro de referencia

---

## ALCANCE

### Instrumento de Medición
- **1 encuesta** de 48 preguntas basada en literatura científica (MBI-GS, Boreout Theory, JD-R Model)
- **5 KPIs** indicadores: Burnout, Boreout, Bienestar, Rotación, Contexto

### Datos Sintéticos (Validación)
- **5 escenarios** organizacionales de referencia (Saludable, Estable, Riesgo Burnout, Riesgo Boreout, Crítico)
- **2.500 empleados por escenario** = **12.500 empleados totales**
- **250 empresas** (50 por escenario)
- **5 CSVs** con resultados (1 por escenario) + CSVs normalizados para Power BI

### Análisis y Visualización
- **3 Jupyter Notebooks** de análisis (EDA, Análisis Profundo, Clustering)
- **6 gráficos comparativos** con explicación e interpretación
- **1 Dashboard interactivo en Power BI** con fórmulas DAX
- **Informe científico** (máximo 4 páginas, formato APA)
- **Presentación PowerPoint** para defensa oral

### Proyección Futura
- Aplicación a **datos reales** de organizaciones
- Generación automática de informes personalizados
- Publicación web con encuesta online

## OBJETIVOS

### Objetivo 1: Diseñar el Instrumento de Medición

Crear una herramienta que permita recabar información de trabajadores y almacenarla de manera estructurada.

**Tareas:**
- Revisión bibliográfica (papers y escalas validadas)
- Definición de la encuesta EBLET (48 preguntas, 7 secciones)
- Validación teórica de dimensiones y constructos
- Documentación del instrumento

**Entregable:** Documento Markdown con la encuesta completa, bases teóricas y mapeo de preguntas.

---

### Objetivo 2: Definir el Modelo Analítico

Desarrollar cálculos, algoritmos y ecuaciones que permitan obtener indicadores para sintetizar y generar estados de trabajadores y empresas a partir de los datos.

**Tareas:**
- Definir los 5 escenarios organizacionales y sus características
- Identificar y definir los 5 KPIs (fórmulas de cálculo)
- Establecer umbrales de clasificación
- Definir efectos organizacionales (cultura, modalidad, departamento)
- Documentar el modelo psicológico

**Entregable:** Documento Markdown con la justificación metodológica completa de parámetros.

---

### Objetivo 3: Generar Datos Sintéticos de Referencia

Crear datos sintéticos realistas para los 5 escenarios como puntos de referencia (benchmarks) que permitan ubicar resultados reales dentro de un espectro.

**Tareas:**
- Implementar funciones generadoras de empresas, empleados y respuestas
- Ajustar parámetros para los 5 escenarios
- Validar que los datos generados coinciden con los escenarios esperados
- Exportar datasets estructurados (CSVs)

**Entregable:**
- Código Python modular (`src/`)
- 5 carpetas con datasets (una por escenario)
- CSVs normalizados para Power BI

---

### Objetivo 4: Análisis Exploratorio y Visualización

Desarrollar elementos visuales (gráficos, informe, dashboard) que permitan la interpretación de los resultados de forma sencilla y accionable.

**Tareas:**
- Definir relaciones entre variables
- Seleccionar visualizaciones idóneas para cada insight
- Justificar qué explica cada gráfico
- Ejecutar análisis estadístico (ANOVA, correlaciones, clustering)
- Construir dashboard en Power BI con fórmulas DAX

**Entregable:**
- 3 Jupyter Notebooks de análisis
- Dashboard Power BI (.pbix)
- Documento con fórmulas DAX

---

### Objetivo 5: Comunicación y Publicación

Publicar herramientas y conclusiones para hacerlas accesibles y facilitar la continuidad del desarrollo/investigación.

**Tareas:**
- Generar repositorio GitHub documentado
- Redactar informe científico (máximo 4 páginas, formato APA)
- Preparar presentación PowerPoint para defensa oral
- Exportar dashboard a PDF

**Entregable:**
- Repositorio GitHub completo
- Informe científico en PDF
- Presentación PowerPoint
- Dashboard exportado en PDF
