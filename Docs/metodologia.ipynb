# 📚 Metodología del Instrumento EBLET

## Encuesta de Bienestar Laboral y Experiencia en el Trabajo


---

## 1. Introducción

La **Encuesta de Bienestar Laboral y Experiencia en el Trabajo (EBLET)** es un instrumento de medición diseñado para evaluar simultáneamente múltiples dimensiones del bienestar organizacional: burnout, boreout, bienestar general, contexto laboral e intención de rotación.

El instrumento se fundamenta en el **modelo Job Demands-Resources (JD-R)** (Demerouti et al., 2001), que postula que el bienestar laboral resulta del equilibrio entre las demandas del trabajo (carga, presión, plazos) y los recursos disponibles (apoyo, autonomía, feedback).

---

## 2. Bases Teóricas por Dimensión

### 2.1 Burnout (Sección D, preguntas 21-29)

**Base teórica:** Maslach Burnout Inventory - General Survey (MBI-GS; Schaufeli et al., 1996)

El burnout se conceptualiza como un síndrome tridimensional:

1. **Agotamiento emocional** (preguntas 21-23): Sensación de falta de energía y fatiga crónica derivada del trabajo.
2. **Cinismo/Despersonalización** (preguntas 24-26): Actitud distante, fría o cínica hacia el trabajo y las personas.
3. **Baja eficacia profesional** (preguntas 27-29): Percepción de incompetencia y disminución del rendimiento.

**Justificación:** El MBI-GS es el instrumento más validado internacionalmente para medir burnout en contextos laborales diversos, con más de 40 años de investigación empírica (Maslach & Leiter, 2016).

**Prevalencia esperada:** Estudios europeos (EUROFOUND, 2018) indican que entre el 15-30% de los trabajadores presentan síntomas significativos de burnout.

---

### 2.2 Boreout (Sección E, preguntas 30-38)

**Base teórica:** Rothlin & Werder (2007); Bianchi et al. (2019)

El boreout es el constructo opuesto al burnout: síndrome derivado de la **infraestimulación crónica** en el trabajo. Se compone de tres dimensiones:

1. **Desinterés** (preguntas 30-32): Falta de motivación, monotonía y ausencia de sentido.
2. **Falta de reto** (preguntas 33-35): Tareas por debajo de las capacidades, sobrecualificación.
3. **Infraocupación y ocultamiento** (preguntas 36-38): Tiempo vacío y presión por fingir estar ocupado.

**Justificación:** Aunque menos estudiado que el burnout, el boreout afecta al 10-20% de trabajadores (Rothlin & Werder, 2007) y tiene consecuencias similares: intención de rotación, depresión, problemas de salud.

**Nota:** El boreout es particularmente relevante en la era post-pandemia, donde el trabajo remoto puede exacerbar la sensación de aislamiento y falta de estímulo.

---

### 2.3 Bienestar y Autoeficacia (Sección F, preguntas 39-45)

**Base teórica:** WHO-5 Well-Being Index (Topp et al., 2015) + Bandura (1997)

Se evalúan dos constructos complementarios:

1. **Satisfacción laboral** (preguntas 39-42): Evaluación cognitiva y afectiva del trabajo.
2. **Autoeficacia laboral** (preguntas 43-45): Creencia en la propia capacidad para resolver problemas y alcanzar objetivos (Bandura, 1997).

**Justificación:** El bienestar no es solo ausencia de malestar (burnout/boreout), sino presencia de factores positivos. La autoeficacia es un predictor clave de rendimiento y resiliencia ante el estrés (Luthans et al., 2007).

---

### 2.4 Contexto Organizacional (Sección C, preguntas 6-20)

**Base teórica:** Job Demands-Resources Model (Demerouti et al., 2001)

Se evalúan cinco dimensiones del contexto laboral:

1. **Apoyo del responsable** (preguntas 6-8): Feedback, ayuda y respaldo del supervisor.
2. **Equilibrio vida-trabajo** (preguntas 9-11): Capacidad de desconectar y compatibilidad con vida personal.
3. **Autonomía** (preguntas 12-14): Control sobre organización, priorización y decisiones.
4. **Apoyo social** (preguntas 15-17): Integración en el equipo y ambiente laboral.
5. **Presión** (preguntas 18-20): Plazos ajustados, múltiples demandas, carga elevada.

**Justificación:** Estas variables son los **recursos y demandas** del modelo JD-R. Los recursos (apoyo, autonomía) protegen contra el burnout; las demandas (presión, carga) lo incrementan (Bakker & Demerouti, 2007).

---

### 2.5 Intención de Rotación (Sección G, preguntas 46-48)

**Base teórica:** Modelo de Mobley (1977)

La intención de rotación es el **predictor más consistente** de la rotación efectiva (Rubio et al., 2015). Se mide mediante tres ítems que capturan:

- Búsqueda activa de empleo
- Probabilidad de cambio a corto plazo
- Disposición ante oportunidades externas

**Justificación:** Medir la rotación efectiva requiere seguimiento longitudinal (1-2 años). La intención de rotación es un proxy válido y ampliamente utilizado en investigación organizacional.

## 3. Proceso de Adaptación de Escalas

### 3.1 Escalas originales utilizadas

| Constructo | Escala original | Autores | Ítems originales |
|------------|----------------|---------|------------------|
| Burnout | MBI-GS | Schaufeli et al., 1996 | 16 ítems |
| Boreout | Boreout Scale | Rothlin & Werder, 2007 | 12 ítems |
| Bienestar | WHO-5 | Topp et al., 2015 | 5 ítems |
| Autoeficacia | WSES | Bandura, 1997 | 8 ítems |
| Rotación | Turnover Intention | Mobley, 1977 | 3 ítems |

### 3.2 Criterios de adaptación

Para reducir la encuesta a 48 preguntas (tiempo de respuesta: 10-15 minutos), se aplicaron los siguientes criterios:

1. **Selección de ítems con mayor carga factorial** en estudios previos.
2. **Mantener representación de todas las subdimensiones** (ej: 3 ítems por cada dimensión del burnout).
3. **Consistencia en el formato de respuesta** (escala Likert 1-5 en todas las secciones).
4. **Lenguaje claro y accesible** para trabajadores de diversos sectores.

### 3.3 Cambios realizados

- **Burnout:** De 16 a 9 ítems (selección de los 3 más representativos por dimensión).
- **Boreout:** De 12 a 9 ítems (manteniendo las 3 subdimensiones).
- **Bienestar:** De 5 a 4 ítems (WHO-5 reducido).
- **Autoeficacia:** De 8 a 3 ítems (selección de los más predictivos).
- **Contexto:** De 20 a 15 ítems (agrupación de dimensiones similares).

## 4. Validación Psicométrica Esperada

### 4.1 Fiabilidad (Consistencia interna)

Se espera que cada dimensión alcance un **Alfa de Cronbach ≥ 0.70**, umbral aceptable en investigación social (Nunnally, 1978):

| Dimensión | Ítems | α esperado |
|-----------|-------|------------|
| Burnout total | 9 | 0.80-0.85 |
| Burnout - Agotamiento | 3 | 0.70-0.75 |
| Burnout - Cinismo | 3 | 0.70-0.75 |
| Burnout - Ineficacia | 3 | 0.65-0.70 |
| Boreout total | 9 | 0.80-0.85 |
| Bienestar | 7 | 0.75-0.80 |
| Contexto | 15 | 0.85-0.90 |
| Rotación | 3 | 0.70-0.75 |

### 4.2 Validez de constructo

Se realizará **análisis factorial exploratorio (EFA)** para verificar que:

1. Los ítems de burnout cargan en un factor común.
2. Los ítems de boreout cargan en un factor distinto.
3. Burnout y boreout están correlacionados negativamente con bienestar.
4. El modelo de 5 factores (burnout, boreout, bienestar, contexto, rotación) explica >60% de la varianza.

### 4.3 Validez convergente y discriminante

- **Convergente:** Burnout correlaciona positivamente con presión (Sección C, preguntas 18-20).
- **Discriminante:** Burnout y boreout correlacionan negativamente entre sí (r < -0.3).

## 5. Limitaciones del Instrumento

### 5.1 Limitaciones metodológicas

1. **Autoinforme:** Las respuestas pueden estar sujetas a sesgos de deseabilidad social o fatiga.
2. **Sección transversal:** No permite inferir causalidad (ej: ¿el burnout causa rotación o viceversa?).
3. **Reducción de ítems:** Al adaptar escalas validadas, se pierde parte de la riqueza psicométrica original.

### 5.2 Limitaciones contextuales

1. **Validación cultural:** El instrumento se ha diseñado en español, pero no se ha validado cross-culturalmente.
2. **Sectores específicos:** Algunas preguntas pueden no ser igualmente relevantes en todos los sectores (ej: "plazos ajustados" en academia vs. ventas).
3. **Modalidad de trabajo:** El impacto del trabajo remoto en boreout/burnout es un área emergente con poca literatura consolidada.

### 5.3 Próximos pasos para validación

1. **Estudio piloto** con N=100-200 trabajadores para calcular fiabilidad real.
2. **Análisis factorial confirmatorio (CFA)** para validar la estructura teórica.
3. **Validación concurrente** comparando con MBI-GS completo en una submuestra.
4. **Estudio longitudinal** para predecir rotación efectiva a 12 meses.

## 6. Aplicación Práctica

### 6.1 Administración

- **Tiempo estimado:** 10-15 minutos
- **Formato:** Online (Google Forms, Typeform, SurveyMonkey) o papel
- **Frecuencia recomendada:** Anual o semestral
- **Muestra mínima:** 30 empleados por organización (para análisis estadístico)

### 6.2 Interpretación de resultados

Los KPIs se calculan como **media de las respuestas** en cada dimensión (escala 1-5):

| KPI | Rango | Interpretación |
|-----|-------|----------------|
| Burnout | 1-2.5 | Bajo |
| | 2.5-3.5 | Moderado |
| | 3.5-5 | Alto (riesgo) |
| Boreout | 1-2.5 | Bajo |
| | 2.5-3.5 | Moderado |
| | 3.5-5 | Alto (riesgo) |
| Bienestar | 1-2.5 | Bajo |
| | 2.5-3.5 | Moderado |
| | 3.5-5 | Alto |

### 6.3 Umbrales de acción

- **Burnout ≥ 3.5:** Intervención urgente (reducción de carga, apoyo psicológico).
- **Boreout ≥ 3.5:** Rediseño de puesto (job crafting, nuevos retos).
- **Bienestar ≤ 2.5:** Revisión general de clima y cultura organizacional.
- **Rotación ≥ 3.5:** Entrevista de permanencia, revisión salarial.
